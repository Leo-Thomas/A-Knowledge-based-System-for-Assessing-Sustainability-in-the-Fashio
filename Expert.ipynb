```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
```

## Economic measures

### Antecedent

```python
#FIS1
material_cost=ctrl.Antecedent(np.arange(0,11,1), 'Material cost')
labour_cost=ctrl.Antecedent(np.arange(0,11,1), 'Labour cost')
#FIS2
lead_time=ctrl.Antecedent(np.arange(0,11,1), 'Lead time')
on_time_delivery=ctrl.Antecedent(np.arange(0,11,1), 'On-time delivery')
#FIS4
product_quality=ctrl.Antecedent(np.arange(0,11,1), 'Product quality')
#FIS31fis1_eco_out = ctrl.Antecedent(np.arange(0,11,1), 'Fis1 eco out')
fis2_eco_out = ctrl.Antecedent(np.arange(0,11,1), 'Fis2 eco out')
#FIS4
fis3_eco_out = ctrl.Antecedent(np.arange(0,11,1), 'Fis3 eco out')
```

### Membership functions

```python
#FIS1
material_cost.automf(3)
labour_cost.automf(3)

#FIS2
lead_time.automf(3)
on_time_delivery.automf(3)

#FIS4
product_quality.automf(3)

#M
fis2_eco_out.automf(3)
fis3_eco_out.automf(3)
```

### Consequent

```python
#FIS1_ECO
fis1_eco = ctrl.Consequent(np.arange(0,11,1),'FIS1_ECO')
fis1_eco['poor'] = fuzz.trimf(fis1_eco.universe, [0,0,5])
fis1_eco['average'] = fuzz.trimf(fis1_eco.universe, [0,5,10])
fis1_eco['good'] = fuzz.trimf(fis1_eco.universe, [5,10,10])

#FIS2_ECO
fis2_eco = ctrl.Consequent(np.arange(0,11,1),'FIS2_ECO')
fis2_eco['poor'] = fuzz.trimf(fis2_eco.universe, [0,0,5])
fis2_eco['average'] = fuzz.trimf(fis2_eco.universe, [0,5,10])
fis2_eco['good'] = fuzz.trimf(fis2_eco.universe, [5,10,10])

#FIS3_ECO
fis3_eco = ctrl.Consequent(np.arange(0,11,1),'FIS3_ECO')
fis3_eco['poor'] = fuzz.trimf(fis3_eco.universe, [0,0,5])
fis3_eco['average'] = fuzz.trimf(fis3_eco.universe, [0,5,10])
fis3_eco['good'] = fuzz.trimf(fis3_eco.universe, [5,10,10])

#FIS4_ECO
fis4_eco = ctrl.Consequent(np.arange(0,11,1),'FIS4_ECO')
fis4_eco['poor'] = fuzz.trimf(fis4_eco.universe, [0,0,5])
fis4_eco['average'] = fuzz.trimf(fis4_eco.universe, [0,5,10])
fis4_eco['good'] = fuzz.trimf(fis4_eco.universe, [5,10,10])
```

```python
#FIS3_ECO
rule1fis3_eco = ctrl.Rule(fis1_eco_out['poor'] & fis2_eco_out['poor'], fis3_eco['poor'])
rule2fis3_eco = ctrl.Rule(fis1_eco_out['average'] & fis2_eco_out['poor'], fis3_eco['poor'])
rule3fis3_eco = ctrl.Rule(fis1_eco_out['good'] & fis2_eco_out['poor'], fis3_eco['average'])
rule4fis3_eco = ctrl.Rule(fis1_eco_out['poor'] & fis2_eco_out['average'], fis3_eco['poor'])
rule5fis3_eco = ctrl.Rule(fis1_eco_out['average'] & fis2_eco_out['average'], fis3_eco['average'])
rule6fis3_eco = ctrl.Rule(fis1_eco_out['good'] & fis2_eco_out['average'], fis3_eco['average'])
rule7fis3_eco = ctrl.Rule(fis1_eco_out['poor'] & fis2_eco_out['good'], fis3_eco['average'])
rule8fis3_eco = ctrl.Rule(fis1_eco_out['average'] & fis2_eco_out['good'], fis3_eco['average'])
rule9fis3_eco = ctrl.Rule(fis1_eco_out['good'] & fis2_eco_out['good'], fis3_eco['good'])

#FIS4_ECO
rule1fis4_eco = ctrl.Rule(product_quality['poor'] & fis3_eco_out['poor'], fis4_eco['poor'])
rule2fis4_eco = ctrl.Rule(product_quality['average'] & fis3_eco_out['poor'], fis4_eco['poor'])
rule3fis4_eco = ctrl.Rule(product_quality['good'] & fis3_eco_out['poor'], fis4_eco['average'])
rule4fis4_eco = ctrl.Rule(product_quality['poor'] & fis3_eco_out['average'], fis4_eco['poor'])
rule5fis4_eco = ctrl.Rule(product_quality['average'] & fis3_eco_out['average'], fis4_eco['average'])
rule6fis4_eco = ctrl.Rule(product_quality['good'] & fis3_eco_out['average'], fis4_eco['average'])
rule7fis4_eco = ctrl.Rule(product_quality['poor'] & fis3_eco_out['good'], fis4_eco['average'])
rule8fis4_eco = ctrl.Rule(product_quality['average'] & fis3_eco_out['good'], fis4_eco['average'])
rule9fis4_eco = ctrl.Rule(product_quality['good'] & fis3_eco_out['good'], fis4_eco['good'])

```

### Consequent

```python
#FIS1_ENV
fis1_env= ctrl.Consequent(np.arange(0,11,1),'FIS1_ENV')
fis1_env['poor'] = fuzz.trimf(fis1_env.universe, [0,0,5])
fis1_env['average'] = fuzz.trimf(fis1_env.universe, [0,5,10])
fis1_env['good'] = fuzz.trimf(fis1_env.universe, [5,10,10])

#FIS2_ENV
fis2_env = ctrl.Consequent(np.arange(0,11,1),'FIS2_ENV')
fis2_env['poor'] = fuzz.trimf(fis2_env.universe, [0,0,5])
fis2_env['average'] = fuzz.trimf(fis2_env.universe, [0,5,10])
fis2_env['good'] = fuzz.trimf(fis2_env.universe, [5,10,10])

#FIS3_ENV
fis3_env = ctrl.Consequent(np.arange(0,11,1),'FIS3_ENV')
fis3_env['poor'] = fuzz.trimf(fis3_env.universe, [0,0,5])
fis3_env['average'] = fuzz.trimf(fis3_env.universe, [0,5,10])
fis3_env['good'] = fuzz.trimf(fis3_env.universe, [5,10,10])

#FIS4_ENV
fis4_env = ctrl.Consequent(np.arange(0,11,1),'FIS4_ENV')
fis4_env['poor'] = fuzz.trimf(fis4_env.universe, [0,0,5])
fis4_env['average'] = fuzz.trimf(fis4_env.universe, [0,5,10])
fis4_env['good'] = fuzz.trimf(fis4_env.universe, [5,10,10])

#FIS4_ENV
fis5_env = ctrl.Consequent(np.arange(0,11,1),'FIS5_ENV')
fis5_env['poor'] = fuzz.trimf(fis5_env.universe, [0,0,5])
fis5_env['average'] = fuzz.trimf(fis5_env.universe, [0,5,10])
fis5_env['good'] = fuzz.trimf(fis5_env.universe, [5,10,10])
```

### Rules

```python
#FIS1_ENV
rule1fis1_env = ctrl.Rule(material_usage['poor'] & recicled_material_usage['poor'], fis1_env['poor'])
rule2fis1_env = ctrl.Rule(material_usage['average'] & recicled_material_usage['poor'], fis1_env['poor'])
rule3fis1_env = ctrl.Rule(material_usage['good'] & recicled_material_usage['poor'], fis1_env['average'])
rule4fis1_env = ctrl.Rule(material_usage['poor'] & recicled_material_usage['average'], fis1_env['poor'])
rule5fis1_env = ctrl.Rule(material_usage['average'] & recicled_material_usage['average'], fis1_env['average'])
rule6fis1_env = ctrl.Rule(material_usage['good'] & recicled_material_usage['average'], fis1_env['average'])
rule7fis1_env = ctrl.Rule(material_usage['poor'] & recicled_material_usage['good'], fis1_env['average'])
rule8fis1_env = ctrl.Rule(material_usage['average'] & recicled_material_usage['good'], fis1_env['average'])
rule9fis1_env = ctrl.Rule(material_usage['good'] & recicled_material_usage['good'], fis1_env['good'])

#FIS2_ENV
rule1fis2_env = ctrl.Rule(water_usage['poor'] & energy_usage['poor'], fis2_env['poor'])
rule2fis2_env = ctrl.Rule(water_usage['average'] & energy_usage['poor'], fis2_env['poor'])
rule3fis2_env = ctrl.Rule(water_usage['good'] & energy_usage['poor'], fis2_env['average'])
rule4fis2_env = ctrl.Rule(water_usage['poor'] & energy_usage['average'], fis2_env['poor'])
rule5fis2_env = ctrl.Rule(water_usage['average'] & energy_usage['average'], fis2_env['average'])
rule6fis2_env = ctrl.Rule(water_usage['good'] & energy_usage['average'], fis2_env['average'])
rule7fis2_env = ctrl.Rule(water_usage['poor'] & energy_usage['good'], fis2_env['average'])
rule8fis2_env = ctrl.Rule(water_usage['average'] & energy_usage['good'], fis2_env['average'])
rule9fis2_env = ctrl.Rule(water_usage['good'] & energy_usage['good'], fis2_env['good'])

#FIS3_ENV = FIS2_ENV RESULT
rule1fis3_env = ctrl.Rule(waste['poor'] & emissions['poor'], fis3_env['poor'])
rule2fis3_env = ctrl.Rule(waste['average'] & emissions['poor'], fis3_env['poor'])
rule3fis3_env = ctrl.Rule(waste['good'] & emissions['poor'], fis3_env['average'])
rule4fis3_env = ctrl.Rule(waste['poor'] & emissions['average'], fis3_env['poor'])
rule5fis3_env = ctrl.Rule(waste['average'] & emissions['average'], fis3_env['average'])
rule6fis3_env = ctrl.Rule(waste['good'] & emissions['average'], fis3_env['average'])
rule7fis3_env = ctrl.Rule(waste['poor'] & emissions['good'], fis3_env['average'])
rule8fis3_env = ctrl.Rule(waste['average'] & emissions['good'], fis3_env['average'])
rule9fis3_env = ctrl.Rule(waste['good'] & emissions['good'], fis3_env['good'])

#FIS4_ENV = FIS1_ENV + FIS2_ENV
rule1fis4_env = ctrl.Rule(fis1_env_out['poor'] & fis2_env_out['poor'], fis4_env['poor'])
rule2fis4_env = ctrl.Rule(fis1_env_out['average'] & fis2_env_out['poor'], fis4_env['poor'])
rule3fis4_env = ctrl.Rule(fis1_env_out['good'] & fis2_env_out['poor'], fis4_env['average'])
rule4fis4_env = ctrl.Rule(fis1_env_out['poor'] & fis2_env_out['average'], fis4_env['poor'])
rule5fis4_env = ctrl.Rule(fis1_env_out['average'] & fis2_env_out['average'], fis4_env['average'])
rule6fis4_env = ctrl.Rule(fis1_env_out['good'] & fis2_env_out['average'], fis4_env['average'])
rule7fis4_env = ctrl.Rule(fis1_env_out['poor'] & fis2_env_out['good'], fis4_env['average'])
rule8fis4_env = ctrl.Rule(fis1_env_out['average'] & fis2_env_out['good'], fis4_env['average'])
rule9fis4_env = ctrl.Rule(fis1_env_out['good'] & fis2_env_out['good'], fis4_env['good'])


#FIS5_ENV = FIS4_ENV + FIS3_ENV
rule1fis5_env = ctrl.Rule(fis4_env_out['poor'] & fis3_env_out['poor'], fis5_env['poor'])
rule2fis5_env = ctrl.Rule(fis4_env_out['average'] & fis3_env_out['poor'], fis5_env['poor'])
rule3fis5_env = ctrl.Rule(fis4_env_out['good'] & fis3_env_out['poor'], fis5_env['average'])
rule4fis5_env = ctrl.Rule(fis4_env_out['poor'] & fis3_env_out['average'], fis5_env['poor'])
rule5fis5_env = ctrl.Rule(fis4_env_out['average'] & fis3_env_out['average'], fis5_env['average'])
rule6fis5_env = ctrl.Rule(fis4_env_out['good'] & fis3_env_out['average'], fis5_env['average'])
rule7fis5_env = ctrl.Rule(fis4_env_out['poor'] & fis3_env_out['good'], fis5_env['average'])
rule8fis5_env = ctrl.Rule(fis4_env_out['average'] & fis3_env_out['good'], fis5_env['average'])
rule9fis5_env = ctrl.Rule(fis4_env_out['good'] & fis3_env_out['good'], fis5_env['good'])
```
